```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.linalg import norm
from collections import Counter
from time import time
from matplotlib import collections  as mc
from sympy import primerange

sns.set_style('whitegrid')
```

```python
df = pd.read_csv("../data/raw/cities.csv")
df.head()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(df.X, df.Y, s=1)
plt.scatter(df.iloc[0: 1, 1], df.iloc[0: 1, 2], s=10, c="red")
plt.grid(False)
plt.show()
```

```python
primes = list(primerange(0, max(df.CityId)))
df['prime'] = df.CityId.isin(primes).astype(int)
```

```python
print(len(df))
print(len(df[df['prime'] == 1]))
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(df[df['prime'] == 0].X, df[df['prime'] == 0].Y, s=1, alpha=0.4)
plt.scatter(df[df['prime'] == 1].X, df[df['prime'] == 1].Y, s=3, alpha=0.6, c='red')
plt.scatter(df.iloc[0: 1, 1], df.iloc[0: 1, 2], s=15, c="green")
plt.grid(False)
plt.title('Visualisation of cities')
plt.show()
```
